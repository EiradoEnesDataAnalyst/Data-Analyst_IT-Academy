```python
# General functions to clean, inspect and transform datasets

# Clean column names
def clean_columns(df):
    df.columns = (
        df.columns
        .str.lower()
        .str.strip()
        .str.replace(" ", "_"))
    return df


# Save_df
def save_df(df, name, folder):
    path = fr"{folder}\{name}.csv"
    df.to_csv(path, index=False, encoding="utf-8-sig")


# Inspect df
def inspect_df(df):
    print("\n------ SHAPE ------")
    print(df.shape)

    print("\n------- NULLS ------")
    print(df.isna().sum().sort_values(ascending=False).head(150))

    print("\n------ DUPLICATES ------")
    print(df.duplicated().sum())

    print("\n------ DATA TYPES ------")
    print(df.dtypes.value_counts())

    print("\n------ UNIQUE VALUES ------")
    print(df.nunique().sort_values(ascending=False).head(150))

    print("\n------ SAMPLE ------")
    display(df.sample(n=min(5, len(df))))
```

<h2 style="text-align:center; color:red; font-size:22px;">
Advanced Time‑Series Completion and Imputation Pipeline
</h2>

<h3 style="text-align:center; color:grey; font-size:18px;">
Key Concepts of the Imputation & Forecasting Pipeline:

1. Forecast Models for Missing Data
The code uses two prediction strategies: a standard linear regression model and a special exponential‑trend model for solar and wind, capped by a maximum growth limit.

2. Vectorized Reconstruction of Full Year Series
All countries are expanded into a complete 2000–2024 timeline using a MultiIndex scaffold, ensuring every year exists even if originally missing.

3. Intelligent Gap Detection
Missing segments are identified through vectorized diff logic, distinguishing internal gaps (filled with previous values) from start/end gaps (requiring model predictions).

4. Recent‑Years‑Only Training Logic
To avoid unrealistic long‑term extrapolation, predictions are trained only on the last 5 valid years when available.

5. Physical & Economic Clamping Rules
Imputed values are constrained by domain‑specific rules: GDP and population have bounded growth/drop rates, shares are clipped to 0–100, and all imputed values are prevented from becoming negative.

6. Automatic Recalculation of Percentage Change Columns
Any column ending in _cons_change_pct is recomputed using grouped percentage change, ensuring consistency after imputation.

</div> 
</h3>

```python
# === FORECAST MODELS ===
def linear_pred(valid_years, valid_values, target_years):
    model = LinearRegression().fit(valid_years.reshape(-1, 1), valid_values)
    return model.predict(target_years.reshape(-1, 1))

def solar_wind_pred(valid_years, valid_values, target_years, max_growth):
    if target_years.max() < 2015:
        return linear_pred(valid_years, valid_values, target_years)

    mask_2015 = valid_years >= 2015
    if mask_2015.sum() >= 2:
        X = valid_years[mask_2015].reshape(-1, 1)
        y = valid_values[mask_2015]

        if np.any(y <= 0):
            pred = linear_pred(valid_years, valid_values, target_years)
        else:
            model = LinearRegression().fit(X, np.log(y))
            pred = np.exp(model.predict(target_years.reshape(-1, 1)))

        last_val = valid_values[-1]
        pred = np.minimum(pred, last_val * max_growth)
        return pred

    return linear_pred(valid_years, valid_values, target_years)


def impute_energy_series(df, start_year=2000, end_year=2024): 
    # Defensive copy
    df = df.copy()

    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if "year" in numeric_cols:
        numeric_cols.remove("year")

    # === GLOBAL PARAMETERS ===
    RECENT_YEARS = 5
    MAX_NAN_RATIO = 0.30     
    MAX_GROWTH = 1.6         
    MAX_GROWTH_GDP = 1.15    
    MAX_GROWTH_POP = 1.04    
    MAX_GDP_DROP = 0.90      
    MAX_POP_DROP = 0.98      

    # === VECTORIZED SCAFFOLDING (Insert missing years) ===
    countries = df["country"].unique()
    years = np.arange(start_year, end_year + 1)
    
    multi_idx = pd.MultiIndex.from_product([countries, years], names=["country", "year"])
    df = df.set_index(["country", "year"]).reindex(multi_idx).reset_index()
    df = df.sort_values(["country", "year"]).reset_index(drop=True)
    
    countries_arr = df["country"].values
    years_arr = df["year"].values

    # Super fast grouped iteration
    grouped = df.groupby("country")
    
    for country, group_indices in grouped.groups.items():
        country_years = years_arr[group_indices]

        for col in numeric_cols:
            series = df[col].values[group_indices].copy()
            nan_mask = np.isnan(series)

            # MAX_NAN_RATIO filter
            if nan_mask.mean() > MAX_NAN_RATIO:
                continue

            valid_idx = np.where(~nan_mask)[0]
            if len(valid_idx) < 2:
                continue

            # VECTORIZED GAP IDENTIFICATION
            is_nan_int = nan_mask.astype(int)
            diffs = np.diff(np.concatenate(([0], is_nan_int, [0])))
            starts = np.where(diffs == 1)[0]
            ends = np.where(diffs == -1)[0]
            
            gaps = [list(range(s, e)) for s, e in zip(starts, ends)]

            valid_years_full = country_years[valid_idx].astype(float)
            valid_values_full = series[valid_idx].astype(float)

            # Use only recent years
            if len(valid_values_full) > RECENT_YEARS:
                valid_years = valid_years_full[-RECENT_YEARS:]
                valid_values = valid_values_full[-RECENT_YEARS:]
            else:
                valid_years = valid_years_full
                valid_values = valid_values_full

            for gap in gaps:
                gap_size = len(gap)
                gap_start = gap[0]
                gap_end = gap[-1]

                is_internal = (gap_start > 0 and gap_end < len(series) - 1)
                is_end = (gap_end == len(series) - 1)

                # INTERNAL GAP
                if is_internal:
                    if gap_size <= 4:
                        prev_val = series[gap_start - 1]
                        if not np.isnan(prev_val):
                            series[gap] = prev_val
                    continue

                # START/END GAP
                target_years = country_years[gap].astype(float)

                if "solar" in col.lower() or "wind" in col.lower():
                    pred = solar_wind_pred(valid_years, valid_values, target_years, MAX_GROWTH)
                else:
                    pred = linear_pred(valid_years, valid_values, target_years)

                # Final trend correction
                if is_end and len(valid_values) >= 2:
                    last_diff = valid_values[-1] - valid_values[-2]
                    for i, p in enumerate(pred):
                        if last_diff < 0 and p > valid_values[-1]:
                            pred[i] = valid_values[-1] + last_diff
                        if last_diff > 0 and p < valid_values[-1]:
                            pred[i] = valid_values[-1] + last_diff

                if gap_size <= 4:
                    series[gap] = pred

            # === PHYSICAL CLAMP APPLIED ONLY TO IMPUTED DATA ===
            # We use `nan_mask` (which remembers what was originally missing)
            # to avoid modifying real historical data.
            col_lower = col.lower()

            if "share" in col_lower:
                series[nan_mask] = np.clip(series[nan_mask], 0, 100)

            elif "pct" in col_lower or "change" in col_lower:
                pass

            elif "gdp" in col_lower:
                for i in range(1, len(series)):
                    if nan_mask[i]:  # Modify ONLY if we imputed it!
                        series[i] = np.clip(series[i], series[i-1] * MAX_GDP_DROP, series[i-1] * MAX_GROWTH_GDP)
                series[nan_mask] = np.clip(series[nan_mask], 0, None)

            elif ("pop" in col_lower) or ("population" in col_lower):
                for i in range(1, len(series)):
                    if nan_mask[i]:  # Modify ONLY if we imputed it!
                        series[i] = np.clip(series[i], series[i-1] * MAX_POP_DROP, series[i-1] * MAX_GROWTH_POP)
                series[nan_mask] = np.clip(series[nan_mask], 0, None)

            else:
                series[nan_mask] = np.clip(series[nan_mask], 0, None)

            df.loc[group_indices, col] = series

    # === RECALCULATE PERCENTAGES ===
    pct_cols = [c for c in df.columns if c.endswith("_cons_change_pct")]

    for col in pct_cols:
        base_col = col.replace("_cons_change_pct", "_consumption")
        if base_col in df.columns:
            calculated_pct = df.groupby("country")[base_col].pct_change() * 100
            df[col] = df[col].fillna(calculated_pct)

    df["year"] = df["year"].astype(int)

    return df
```

```python
# Data Extraction - IRENA 

def extract_irena_data(path):
    """
    Extract all sheets from the IRENA Excel file.

    Args:
        path (str): File path to the Excel file

    Returns:
        dict: Dictionary with sheet names as keys and DataFrames as values
    """
    data = {}

    try:
        excel_file = pd.ExcelFile(path)
        
        for sheet in excel_file.sheet_names:
            df = pd.read_excel(excel_file, sheet_name=sheet)
            data[sheet] = df

        print(f"IRENA file loaded successfully ({len(data)} sheets)")
        return data

    except Exception as e:
        print(f"Error loading IRENA file: {e}")
        raise
```

```python
def transform_irena_capacity(Country_Irena):
    """
    Clean, normalize, map, aggregate and transform IRENA installed capacity data
    into OWID-style categories.

    Returns:
        df_mapped (DataFrame): cleaned and mapped dataset (before pivot)
        Irena_Transformed (DataFrame): aggregated dataset (after pivot)
    """

    # 1. Full normalization: strip spaces, lowercase, unify ISO3 format
    for col in ["producer_type", "technology", "iso3_code"]:
        Country_Irena[col] = (
            Country_Irena[col]
            .astype(str)
            .str.strip()
            .str.lower()
        )

    Country_Irena["iso3_code"] = Country_Irena["iso3_code"].str.upper()

    # 2. Updated mapping (based on full technology names)
    mapping = {
        "solar energy": "solar",
        "solar photovoltaic": "solar",
        "solar thermal energy": "solar",

        "onshore wind energy": "wind",
        "offshore wind energy": "wind",
        "wind energy": "wind",

        "renewable hydropower": "hydro",
        "mixed hydro plants": "hydro",
        "hydropower (excl. pumped storage)": "hydro",

        "nuclear": "nuclear",
        "coal and peat": "coal",

        "natural_gas": "gas",
        "natural gas": "gas",

        "oil": "oil",

        "solid biofuels": "biofuel",
        "biogas": "biofuel",
        "liquid biofuels": "biofuel",
        "renewable municipal waste": "biofuel",

        "geothermal energy": "other_renewables",
        "marine energy": "other_renewables",
        "other renewable energy": "other_renewables",
        "multiple renewables*": "other_renewables",

        "fossil fuels n.e.s.": "other_fossil",
        "non-renewable municipal waste": "other_fossil",
        "other non-renewable energy": "other_fossil"
    }

    # Apply mapping
    df_mapped = Country_Irena.copy()
    df_mapped["owid_category"] = df_mapped["technology"].map(mapping)

    # Remove unmapped rows (e.g., pumped storage)
    df_mapped = df_mapped.dropna(subset=["owid_category"])

    # 3. Pivot transformation WITHOUT filtering producer_type
    Irena_Transformed = df_mapped.pivot_table(
        index=["iso3_code", "country", "year"],
        columns="owid_category",
        values="electricity_installed_capacity_(mw)",
        aggfunc="sum"
    ).reset_index()

    # 4. Cleanup and aggregated totals
    Irena_Transformed.columns.name = None

    cols = [
        c for c in Irena_Transformed.columns
        if c not in ["iso3_code", "country", "year"]
    ]

    Irena_Transformed = Irena_Transformed.rename(
        columns={c: f"{c}_capacity_mw" for c in cols}
    )

    Irena_Transformed = Irena_Transformed.fillna(0)

    # Fossil = coal + gas + oil + other_fossil
    Irena_Transformed["fossil_capacity_mw"] = (
        Irena_Transformed.get("coal_capacity_mw", 0)
        + Irena_Transformed.get("gas_capacity_mw", 0)
        + Irena_Transformed.get("oil_capacity_mw", 0)
        + Irena_Transformed.get("other_fossil_capacity_mw", 0)
    )

    # Renewables = solar + wind + hydro + biofuel + other_renewables
    Irena_Transformed["renewables_capacity_mw"] = (
        Irena_Transformed.get("solar_capacity_mw", 0)
        + Irena_Transformed.get("wind_capacity_mw", 0)
        + Irena_Transformed.get("hydro_capacity_mw", 0)
        + Irena_Transformed.get("biofuel_capacity_mw", 0)
        + Irena_Transformed.get("other_renewables_capacity_mw", 0)
    )

    # Low carbon = renewables + nuclear
    Irena_Transformed["low_carbon_capacity_mw"] = (
        Irena_Transformed["renewables_capacity_mw"]
        + Irena_Transformed.get("nuclear_capacity_mw", 0)
    )

   # Prepare for merging with OWID
    Irena_Transformed = Irena_Transformed.rename(columns={"iso3_code": "iso_code"})

    # Keep only rows where iso_code exists
    Irena_Transformed = Irena_Transformed[Irena_Transformed["iso_code"].notna()]


    print("Transformation completed successfully.")

    return df_mapped, Irena_Transformed
```

```python
# Run tranform

df_mapped, Irena_Transformed = transform_irena_capacity(Country_Irena)
Irena_Transformed.tail(30)
```

```python
# 1. Total installed capacity in the original dataset (after mapping)
total_original = df_mapped["electricity_installed_capacity_(mw)"].sum()

# 2. Identify base capacity columns (exclude aggregated totals)
base_columns = [
    c for c in Irena_Transformed.columns
    if c.endswith("_capacity_mw")
    and c not in [
        "fossil_capacity_mw",
        "renewables_capacity_mw",
        "low_carbon_capacity_mw"]]

# 3. Total installed capacity in the transformed dataset
total_transformed = Irena_Transformed[base_columns].sum().sum()

# 4. Global consistency check
print("\n--- GLOBAL CONSISTENCY CHECK (ALL COUNTRIES) ---")
print(f"Total MW (Original):     {total_original:,.2f}")
print(f"Total MW (Transformed):  {total_transformed:,.2f}")
print(f"Difference:              {total_original - total_transformed:,.2f}")

# 5. Country-level consistency check (ISO3 totals)
country_check = (
    df_mapped.groupby("iso3_code")["electricity_installed_capacity_(mw)"].sum()
    - Irena_Transformed[base_columns]
        .groupby(Irena_Transformed["iso_code"])
        .sum()
        .sum(axis=1)
)

print("\n--- COUNTRY-LEVEL CONSISTENCY CHECK (ISO3) ---")
print(country_check)
```

```python
# =========================================================
# 0. CLEANING & PREPARATION — UPDATED EMBER 2030
# =========================================================

# 1. Keep the oldest target if duplicates exist for the same collection
ember_2030 = (
    ember_2030
    .sort_values("collection_year")
    .drop_duplicates(
        subset=["country_code", "target_year", "fuel_category", "metric"],
        keep="first"))

# 2. Metric handling: keep capacity_total_gw (now part of your variable list)
ember_2030 = ember_2030[
    ~ember_2030["metric"].isin(["capacity_additional_mw"])]

# 3. Merge Onshore and Offshore Wind into a single 'Wind' category
ember_2030["fuel_category"] = ember_2030["fuel_category"].replace({
    "Onshore Wind": "Wind",
    "Offshore Wind": "Wind"})

# 4. Convert values to float (handling comma separators)
ember_2030["value"] = (
    ember_2030["value"]
    .astype(str)
    .str.replace(",", ".", regex=False)
    .astype(float))

# 5. Aggregate duplicates created by merging Wind categories
ember_2030 = (
    ember_2030.groupby(
        ["country_code", "target_year", "fuel_category", "metric", "collection_year"],
        as_index=False
    ).agg({"value": "sum"}))

# =========================================================
# 6. UPDATED MAPPING (aligned with df_pivoted['variable'])
# =========================================================

mapping = {
    # Electricity generation and shares
    ("Renewables", "generation_twh"): "renewables_electricity",
    ("Renewables", "share_of_generation_pct"): "renewables_share_elec",
    ("Clean", "generation_twh"): "low_carbon_electricity",
    ("Clean", "share_of_generation_pct"): "low_carbon_share_elec",
    ("Solar", "generation_twh"): "solar_electricity",
    ("Solar", "share_of_generation_pct"): "solar_share_elec",
    ("Wind", "generation_twh"): "wind_electricity",
    ("Wind", "share_of_generation_pct"): "wind_share_elec",
    ("Hydro", "generation_twh"): "hydro_electricity",
    ("Hydro", "share_of_generation_pct"): "hydro_share_elec",
    ("Nuclear", "generation_twh"): "nuclear_electricity",
    ("Nuclear", "share_of_generation_pct"): "nuclear_share_elec",
    ("Bioenergy", "generation_twh"): "biofuel_electricity",
    ("Bioenergy", "share_of_generation_pct"): "biofuel_share_elec",
    ("Fossil", "generation_twh"): "fossil_electricity",
    ("Fossil", "share_of_generation_pct"): "fossil_share_elec",
    ("Coal", "generation_twh"): "coal_electricity",
    ("Coal", "share_of_generation_pct"): "coal_share_elec",
    ("Gas", "generation_twh"): "gas_electricity",
    ("Gas", "share_of_generation_pct"): "gas_share_elec",
    ("Wind and solar", "generation_twh"): "wind_solar_electricity",
    ("Wind and solar", "share_of_generation_pct"): "wind_solar_share_elec",

    # Installed capacity (GW)
    ("Renewables", "capacity_total_gw"): "renewables_capacity_gw",
    ("Solar", "capacity_total_gw"): "solar_capacity_gw",
    ("Wind", "capacity_total_gw"): "wind_capacity_gw",
    ("Hydro", "capacity_total_gw"): "hydro_capacity_gw",
    ("Nuclear", "capacity_total_gw"): "nuclear_capacity_gw",
    ("Bioenergy", "capacity_total_gw"): "biofuel_capacity_gw",
    ("Fossil", "capacity_total_gw"): "fossil_capacity_gw",
    ("Coal", "capacity_total_gw"): "coal_capacity_gw",
    ("Gas", "capacity_total_gw"): "gas_capacity_gw",
    ("Clean", "capacity_total_gw"): "low_carbon_capacity_gw",

    # Primary energy consumption share
    ("Renewables", "share_of_consumption_pct"): "renewables_share_energy",}

# Build mapping DataFrame
mapping_df = pd.DataFrame(
    [(k[0], k[1], v) for k, v in mapping.items()],
    columns=["fuel_category", "metric", "variable"])

# Merge mapping into Ember dataset
ember_2030 = ember_2030.merge(mapping_df, on=["fuel_category", "metric"], how="left")

# =========================================================
# 7. VALIDATION & FINAL PREPARATION
# =========================================================

# Check for unmapped combinations
unmapped = ember_2030[ember_2030["variable"].isna()][["fuel_category", "metric"]].drop_duplicates()
if not unmapped.empty:
    print("Warning: Unmapped Ember combinations found:")
    print(unmapped)

# Remove unmapped rows and rename for final merge
ember_2030 = (
    ember_2030
    .dropna(subset=["variable"])
    .rename(columns={
        "country_code": "iso_code",
        "value": "2030_target"}))

print(f"Ember dataset updated successfully. Countries included: {ember_2030['iso_code'].nunique()}")
```
